### Explaining Profit using Equations

Profit, a key factor in decision making, can be represented using mathematical equations. For example:

$$
\text{Profit} = \text{Revenue} - \text{Expenses}
$$

The equation \(Profit = Revenue - Expenses\) captures the relationship between profit, revenue \(R\), and expenses \(E\). It quantifies the financial gain obtained by subtracting expenses from revenue.

- Another way to express profit is as a function of revenue (R) and expenses (E):

$$
\text{Profit} = f(R, E)
$$

These equations illustrate how profit depends on revenue and expenses, emphasizing the mathematical relationship and the operations involved.

Dependent and Independent Variables

In a mathematical model, variables can be categorized as dependent or independent. Let's consider the example of profit, but this concept applies to any variable of interest.

- Dependent Variable: The variable whose value depends on other variables is called the dependent variable. In this case, profit is the dependent variable because its value depends on the values of revenue ($R$) and expenses ($E$).

- Independent Variables: The variables that influence or affect the value of the dependent variable are called independent variables. In this case, revenue and expenses are the independent variables as the value of profit depends on these factors.

More generally, if we have a variable of interest, denoted as $Y$, it can be considered the dependent variable. It can be expressed as a function of independent variables, denoted as $X_1, X_2, ..., X_n$. The relationship between the dependent variable and independent variables is captured by a mathematical function or equation.

So, $Y$ represents the dependent variable, and $X_1, X_2, ..., X_n$ represent the independent variables. There is a mathematical operation or function, denoted as $f$, that can be performed based on the values of these independent variables to determine the value of $Y$. The equation can be written as:

$$
Y = f(X_1, X_2, ..., X_n)
$$

In the case of profit, it can be represented as:

$$
\text{Profit} = f(\text{Revenue}, \text{Expenses})
$$

In summary, $Y$ is the dependent variable, and we have $n$ independent variables in this mathematical model. The functional relationship is specified by the function $f$, which defines the operations and relationships between the independent variables $X_1, X_2, ..., X_n$ that determine the value of $Y$.

## Formulation

In the context of mathematical modeling and optimization, a formula plays a crucial role in representing real-world situations or systems. Mathematical modeling involves formulating these formulas to capture the underlying relationships and rules governing a particular scenario. These formulas enable us to describe the problem in mathematical terms, making it amenable to analysis and prediction. Optimization techniques, such as linear programming, are then applied to these formulated models to find the optimal solution that maximizes or minimizes an objective while satisfying the defined constraints. By using formulas and optimization methods, we can effectively tackle complex problems, make informed decisions, and find optimal solutions in various fields of study and application.

#### Common Formulas Used in a Business Context

**Profit Formula** (Linear)
$$\text{Profit} = \text{Revenue} - \text{Cost}$$

**Break-Even Point** (Linear)
$$\text{Break-Even Point} = \frac{\text{Fixed Costs}}{\text{Selling Price per Unit} - \text{Variable Cost per Unit}}$$

**Return on Invenstment - ROI**(Linear)
$$\text{ROI} = \frac{\text{Net Profit}}{\text{Investment Cost}} \times 100$$

**Production Volume Formula** (Quadratic)
$$\text{Production Volume} = -\frac{b}{2a}$$

**Forecasting**(Polynomial) 
$$\text{Sales} = aX^n + bX^{(n-1)} + \ldots + kX^2 + mX + c$$

## [Using Python to Plot Mathematical Models](#using-python-to-plot-mathematical-models)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(-20, 20, 100)

# Define the variables
a = 0.5
b = -1
c = 3

# Define the formulas
profit = a * x + b
break_even_point = c / (a - b)
return_on_investment = (profit / c) * 100
production_volume = -b / (2 * a)
forecasting = a * x**2 + b * x + c

# Plot the formulas
plt.figure(figsize=(10, 6))
plt.plot(x, profit, label='Profit', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.plot(break_even_point, 0, marker='o', markersize=8, color='red', label='Break-Even Point')
plt.plot(x, return_on_investment, label='Return on Investment', color='green')
plt.plot(production_volume, 0, marker='o', markersize=8, color='purple', label='Production Volume')
plt.plot(x, forecasting, label='Forecasting', color='orange')

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Business Formulas')

# Add legend
plt.legend()

# Adjust plot limits
plt.xlim(0, 20)
plt.ylim(-20, 50)

# Display the plot
plt.grid(True)
plt.show()
```

Formulation refers to the process of expressing a problem or concept in a precise and structured manner using mathematical or symbolic notation. In the context of linear optimization, formulation involves defining the objective function, decision variables, and constraints in a mathematical form. It entails representing the relationships and dependencies between variables, expressing the optimization goal, and specifying any limitations or requirements. Formulation essentially converts a real-world problem into a mathematical representation that can be solved using optimization techniques

## Formulation Steps

1. **Identify the Objective**:
   - Clearly define the goal or objective of the optimization problem.

2. **Specify Decision Variables**:
   - Determine the variables that can be controlled and whose values need to be determined.

3. **Formulate the Objective Function**:
   - Express the objective as a mathematical function of the decision variables.

4. **Define Constraints**:
   - Identify any restrictions or limitations on the decision variables.

5. **Consider Hidden Conditions**:
   - Account for any additional conditions or requirements not explicitly stated.

6. **Formulate the Complete Model**:
   - Combine the objective function, decision variables, and constraints into a comprehensive mathematical representation.

7. **Review and Refine**:
   - Double-check the formulation for accuracy and clarity.

# [Example 1: Oak Barrel Brewery](#example-1-oak-barrel-brewery)

In the manufacturing industry, decision-making plays a crucial role in optimizing production processes and maximizing profits. Let's dive into a specific example that illustrates the formulation of a linear optimization problem in the context of a brewery. Imagine a renowned brewery called Oak Barrel, known for producing high-quality ale and beer.

In this scenario, we will explore how the brewery manager can make informed decisions about the quantities of ale and beer to produce, taking into account the available resources, raw materials, and profit margins. By formulating the problem as a linear optimization model, we can assist the brewery manager in determining the optimal production quantities to achieve the highest possible profits.

The decision variables in this problem are:

- (a) Amount of ale to produce

<details>
  <summary>Click to reveal the Answer</summary>
 
 The production plant needs to decide how much ale and beer to produce, so these two should be the decision variables. Specifically, you can write `x = amount of ale to produce`, and `y = amount of beer to produce` in units of barrels.
</details>

1. **Identify the Objective**:
   - Maximize the total profit by determining the quantities of ale and beer to produce.

2. **Specify Decision Variables**:
   - Let X represent the quantity of ale to produce and Y represent the quantity of beer to produce.

3. **Formulate the Objective Function**:
   - The objective function is to maximize the profit, expressed as: Maximize Profit = 13X + 23Y.

4. **Define Constraints**:
   - There are three constraints based on the availability of raw materials:
     - Corn constraint: 5X + 15Y ≤ 480 (units of corn)
     - Hops constraint: 4X + 4Y ≤ 160 (units of hops)
     - Malt constraint: 35X + 20Y ≤ 1190 (units of malt)
   - Additionally, there are non-negativity constraints: X ≥ 0 and Y ≥ 0.

5. **Consider Hidden Conditions**:
   - None in this example.<br>

6. **Formulate the Complete Model**:
   - Combining the objective function and constraints, the linear program can be written as:
     Maximize Profit = 13X + 23Y
     Subject to:
     5X + 15Y ≤ 480
     4X + 4Y ≤ 160
     35X + 20Y ≤ 1190
     X, Y ≥ 0

7. **Review and Refine**:
   - Double-check the formulation to ensure accuracy and clarity.

```python
from pulp import *

# Create the problem variable
prob = LpProblem("Brewery_Profit_Optimization", LpMaximize)

# Define the decision variables
X = LpVariable("X", lowBound=0)  # Quantity of ale to produce
Y = LpVariable("Y", lowBound=0)  # Quantity of beer to produce

# Set up the objective function
prob += 13*X + 23*Y, "Total_Profit"

# Add the constraints
prob += 5*X + 15*Y <= 480, "Corn_Constraint"
prob += 4*X + 4*Y <= 160, "Hops_Constraint"
prob += 35*X + 20*Y <= 1190, "Malt_Constraint"

# Solve the problem
prob.solve()

# Print the optimal solution
print("Optimal Solution:")
print("X =", X.varValue)
print("Y =", Y.varValue)
print("Total Profit =", prob.objective.value())
```

In this code, we use lpSum from pulp to simplify the objective function by using the += operator to add terms to it. It allows us to incrementally build the objective function without explicitly rewriting the entire function. The += operator is equivalent to the statement prob += ... where prob is the problem variable.

```python
X_optimal = 12.0
Y_optimal = 28.0

# Define the range of x values
x = np.linspace(0, 30, 100)

# Define the constraints
corn_constraint = (480 - 5*x) / 15
hops_constraint = (160 - 4*x) / 4
malt_constraint = (1190 - 35*x) / 20

# Calculate the feasible region
feasible_region = np.minimum.reduce([corn_constraint, hops_constraint, malt_constraint])

# Create the figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Left subplot - Feasible Region and Constraints
ax1.plot(x, corn_constraint, 'r', label='Corn Constraint')
ax1.plot(x, hops_constraint, 'g', label='Hops Constraint')
ax1.plot(x, malt_constraint, 'm', label='Malt Constraint')
ax1.set_xlabel('Quantity of Ale (X)')
ax1.set_ylabel('Quantity of Beer (Y)')
ax1.set_title('Constraints')
ax1.legend()
ax1.grid(True)

# Right subplot - Optimal Production Quantities
ax2.plot(x, corn_constraint, 'r', label='Corn Constraint')
ax2.plot(x, hops_constraint, 'g', label='Hops Constraint')
ax2.plot(x, malt_constraint, 'm', label='Malt Constraint')
ax2.fill_between(x, 0, feasible_region, where=(feasible_region >= 0), color='lightblue', alpha=0.5)
ax2.plot(x, feasible_region, 'b', label='Feasible Region')
ax2.plot(X_optimal, Y_optimal, 'ro', label='Optimal Solution')
ax2.set_xlabel('Quantity of Ale (X)')
ax2.set_ylabel('Quantity of Beer (Y)')
ax2.set_title('Optimal Production Quantities')
ax2.legend()
ax2.grid(True)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# Create the problem variable
prob = LpProblem("Brewery_Profit_Optimization", LpMaximize)

# Define the decision variables
X = LpVariable("X", lowBound=0)  # Quantity of ale to produce
Y = LpVariable("Y", lowBound=0)  # Quantity of beer to produce

# Set up the objective function
prob += 13*X + 23*Y, "Total_Profit"

# Add the constraints
prob += 5*X + 15*Y <= 480, "Corn_Constraint"
prob += 4*X + 4*Y <= 160, "Hops_Constraint"
prob += 35*X + 20*Y <= 1190, "Malt_Constraint"

# Add the additional constraint
prob += X >= 10, "Additional_Constraint"

# Solve the problem
prob.solve()

# Print the optimal solution
print("Optimal Solution:")
print("X =", X.varValue)
print("Y =", Y.varValue)
print("Total Profit =", prob.objective.value())
```
