# Linear Regression Part

```python
# This is a convention. Please pay atention to CAPITALIZATION of the variables.
X = iris['petal_length'] # This is a CAPITAL X
y = iris['petal_width']
```

```python
sns.scatterplot(x=X, y=y)
```

```python

serie1 = pd.Series(np.random.uniform(2,3,50))
serie2 = pd.Series(np.random.uniform(7,9,50))

X_test = pd.concat([serie1, serie2], ignore_index=True)
```

```python
model = sm.OLS(y,X).fit()
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
sns.scatterplot(x=X, y=y, color='blue')
sns.scatterplot(x=X_test, y=y_pred, color='red')
plt.title('My first prediction 😎')
```
